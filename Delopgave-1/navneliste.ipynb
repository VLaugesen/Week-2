```python
import matplotlib.pyplot as plt

def perform_analysis(name_list):
    letters = {}
    lengths = {}

    for name in name_list:
        length = len(name)
        lengths[length] = lengths.get(length, 0) + 1
        for letter in name:
            letters[letter] = letters.get(letter, 0) + 1
    return letters, lengths

with open("../Data/Navneliste.txt") as file:
    names = file.read().lower().split(',')

    print("Sorted alphabetically:", sorted(names))
    print("Sorted by length:", sorted(names, key=str.__len__))

    letters, lengths = perform_analysis(names)
    letters_no_dup, lengths_no_dup = perform_analysis(list(set(names)))

print(letters)
```

```python
fig, axs = plt.subplots(3, figsize=(8, 6))
axs[0].bar(letters.keys(), letters.values())

# Sorting the plot alphabetically
xs, ys = zip(*sorted(letters.items()))
axs[1].bar(xs, ys)

axs[2].bar(lengths.keys(), lengths.values())


plt.show()

print(lengths.values())
```

```python
fig, axs = plt.subplots(3, figsize=(8, 6))
axs[0].bar(letters_no_dup.keys(), letters_no_dup.values())

# Sorting the plot alphabetically
xs, ys = zip(*sorted(letters_no_dup.items()))
axs[1].bar(xs, ys)

axs[2].bar(lengths_no_dup.keys(), lengths_no_dup.values())


plt.show()

print(lengths_no_dup.values())
```
