```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../Data/DKHousingPricesSample100k.csv")
print(df.head(10))

def plot_average_by_group(df, group):
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    goupby_region = df.groupby("region")
    groupby_region_mean = goupby_region["purchase_price"].mean()
    print(groupby_region_mean)
    groupby_region_mean.plot(title="Average purchase price", ax=axs[0], kind="bar")


fig, axs = plt.subplots(1, 2, figsize=(12, 5))

goupby_region = df.groupby("region")
groupby_region_mean = goupby_region["purchase_price"].mean()
print(groupby_region_mean)
groupby_region_mean.plot(title="Average purchase price", ax=axs[0], kind="bar")

groupby_house_type = df.groupby("house_type")
groupby_house_type_mean = groupby_house_type["purchase_price"].mean()
print(groupby_house_type_mean)
groupby_house_type_mean.plot(title="Average purchase price", ax=axs[1], kind="bar")

```
